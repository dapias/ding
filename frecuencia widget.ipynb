```python
from IPython.html.widgets import interact
```

```python
def iterar(frecuencia,  usar_mod=True):
    num_total = 7
    reservorio = dg.Reservorio()
    caja = dg.Caja(np.float(num_total - 1.)/2. + 1.)
    lista = dg.crear_particulas_aleatorias(caja.tamano,num_total,frecuencia,reservorio)
    reglas = dg.ReglasColision(caja, reservorio)
    sim = dg.Simulacion(lista, reglas)
    sim.run(0,0,1,numero_puntos = 100)
    
    tiempo = []
    num_particulas = (num_total - 1) * 0.5 + 1
    num_osciladores = (num_total - 1) * 0.5
    px = [[] for _ in xrange(int(num_particulas))]
    osx = [[] for _ in xrange(int(num_osciladores))]
    
    t = 0        
    dt = 0.005
    


    for i in xrange(len(sim.t_eventos) - 1):
    
        while  t < float(sim.t_eventos[i + 1]):
            
            for j in xrange(int(num_particulas)):
                px[j].append(sim.registro_posiciones["Particula" + str(j + 1)][i] + (t - sim.t_eventos[i]) * sim.registro_velocidades["Particula" + str(j + 1)][i])
#
            for j in xrange(int(num_osciladores)):
                osx[j].append(sim.registro_amplitudes["Oscilador" +  str(j + 1) ][i]*np.sin(np.float(frecuencia*(t - sim.t_eventos[i]) + sim.registro_fases["Oscilador" +  str(j + 1)][i])) + sim.osciladores[j].equilibrio)


            tiempo.append(t)
            t += dt

    for j in xrange(int(num_particulas)):
        plt.plot(tiempo, px[j], '--')
    for j in xrange(int(num_osciladores)):
        plt.plot(tiempo, osx[j], '--')
    
    #plt.axis([0,25,-2,2])
    plt.xlabel('t')
    plt.ylabel('x(t)')

    
```

```python
iterar(0.1)
```

```python
interact(iterar,frecuencia =(0.5,5.,0.5))
```
